```python
import mysql.connector
import pandas as pd
```

```python
# Definindo os parâmetros para conxeão do colab com o banco faculdade
host = '34.46.165.135'
user = 'root'
password = ''
database = 'logistica'
```

```python
!curl ipecho.net/plain
```

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco
query = 'show tables;' # Query ou consulta desejada
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
tabelas = pd.DataFrame(result,columns=cursor.column_names) # convertendo
display(tabelas) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco
query = '''
SELECT *
FROM Motoristas;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
motoristas = pd.DataFrame(result,columns=cursor.column_names) # convertendo
display(motoristas) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```
