```python
import pandas as pd
```

```python
bp = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQoh3kZ98EYa9Dq3BTKx6b9WBZfQlTYycRIytxdWS6ycIuGh_cndGoaWz-P2eORmqBpKmDzVbZkvBQJ/pub?gid=439773759&single=true&output=csv',header=1)
bp.head()
```

```python
bp = bp.transpose()
pd.set_option('display.max_columns',None) #Código para visualizar todas as colunas
bp.head(4)
```

```python
dre = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQoh3kZ98EYa9Dq3BTKx6b9WBZfQlTYycRIytxdWS6ycIuGh_cndGoaWz-P2eORmqBpKmDzVbZkvBQJ/pub?gid=1719424801&single=true&output=csv',header=1)
dre
```

```python
dre = dre.transpose()
pd.set_option('display.max_columns',None)
dre.head(4)
```

```python
bpr = bp.iloc[:,[4,5,27]]
bpr.head()
```

```python
bpr = bpr.rename(columns={4:'clientes',5:'estoques',27:'fornecedores'})
bpr.head()
```

```python
bpr = bpr.drop(bpr.index[[0]])
bpr.head()
```

```python
drer = dre.iloc[:,[2,3]]
drer = drer.rename(columns={2:'receita_liquida',3:'custo'})
drer = drer.drop(drer.index[[0]])
drer.head()
```

```python
dfi = bpr
dfi
```

```python
dfi['receita_liquida'] = drer['receita_liquida']
dfi['custo'] = drer['custo']
dfi.head()
```

```python
dfi['pme'] = abs((dfi['estoques']/dfi['custo'])*90)
dfi['pmr'] = abs((dfi['clientes']/dfi['receita_liquida'])*90)
dfi['pmp'] = abs((dfi['fornecedores']/dfi['custo'])*90)
dfi['co'] = dfi['pme'] + dfi['pmr']
dfi['cc'] = dfi['co'] - dfi['pmp']
dfi.head()
```
