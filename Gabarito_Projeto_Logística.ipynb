# Questã 01 - Qual motorista tem a maior quantidade de pedidos entregues no menor tempo?

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco
query = '''
SELECT
    Motorista,
    COUNT(ID_Pedido) AS Total_Pedidos,
    AVG(TIMESTAMPDIFF(HOUR, Data_Pedido, Data_Entrega_Real)) AS Tempo_Medio_Entrega_Horas
FROM
    Pedidos
GROUP BY
    Motorista
ORDER BY
    Tempo_Medio_Entrega_Horas ASC, Total_Pedidos DESC;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
quest1 = pd.DataFrame(result,columns=cursor.column_names) # convertendo
display(quest1) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco
query = '''
SELECT
    M.Nome AS Nome_Motorista,
    COUNT(P.ID_Pedido) AS Total_Pedidos,
    AVG(TIMESTAMPDIFF(HOUR, P.Data_Pedido, P.Data_Entrega_Real)) AS Tempo_Total_Entrega_Horas
FROM
    Pedidos P
JOIN
    Motoristas M ON P.Motorista = M.ID_Motorista
GROUP BY
    M.Nome
ORDER BY
    Total_Pedidos DESC, Tempo_Total_Entrega_Horas ASC;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
quest1a = pd.DataFrame(result,columns=cursor.column_names) # convertendo
display(quest1a) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

# Questão 2 - Existe uma correlação entre a avaliação de desempenho dos motoristas e o tempo médio de entrega?

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco
query = '''
SELECT
    M.Avaliacao_Desempenho,
    AVG(TIMESTAMPDIFF(HOUR, P.Data_Pedido, P.Data_Entrega_Real)) AS Tempo_Medio_Entrega_Horas
FROM
    Pedidos P
JOIN
    Motoristas M ON P.Motorista = M.ID_Motorista
GROUP BY
    M.Avaliacao_Desempenho
ORDER BY
    M.Avaliacao_Desempenho DESC;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
quest2 = pd.DataFrame(result,columns=cursor.column_names) # convertendo
display(quest2) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

# Questão 3 - Quais regiões têm os maiores custos de transporte por quilômetro?

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco
query = '''
SELECT
    Regiao,
    SUM(Custo_Transporte) / SUM(Distancia_km) AS Custo_por_KM
FROM
    Pedidos
GROUP BY
    Regiao
ORDER BY
    Custo_por_KM DESC;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
quest3 = pd.DataFrame(result,columns=cursor.column_names) # convertendo
display(quest3) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```
