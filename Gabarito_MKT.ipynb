```python
'''
Verificar a função SQL - By Rodrigo apud Chat GPT
'''

'''
Existe uma correlação entre o investimento total em campanhas e o número de
impressões geradas?
'''
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
SELECT
    (COUNT(*) * SUM(investimento_total * Impressoes) - SUM(investimento_total) * SUM(Impressoes)) /
    (SQRT(COUNT(*) * SUM(investimento_total * investimento_total) - SUM(investimento_total) * SUM(investimento_total)) *
     SQRT(COUNT(*) * SUM(Impressoes * Impressoes) - SUM(Impressoes) * SUM(Impressoes))
    ) AS correlacao
FROM
    Campanhas;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
q3 = pd.DataFrame(result,columns=cursor.column_names) # convertendo a variável result em um df pandas chamado tabelas
display(q3) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
'''
Sugestões de gráficos: Série temporal (linha)
'''
'''
 Como a duração da campanha influencia o número de cliques e conversões?
'''
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
SELECT DATEDIFF(data_fim, data_inicio) AS duracao_campanha,
       SUM(cliques) AS total_cliques,
       SUM(conversoes) AS total_conversoes
FROM Campanhas
GROUP BY duracao_campanha
ORDER BY duracao_campanha;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
tabelas = pd.DataFrame(result,columns=cursor.column_names) # convertendo a variável result em um df pandas chamado tabelas
display(tabelas) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
'''
Sugestões de gráficos: Barras empilhada, ou colunas empilhadas
'''
'''
Quais são as principais diferenças de desempenho entre campanhas veiculadas no
Google e no Instagram?
'''
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
SELECT plataforma,
       AVG(Cliques/Impressoes) AS media_taxa_cliques,
       AVG(Conversoes/Cliques) AS media_taxa_conversao
FROM Campanhas
WHERE plataforma IN ('Google', 'Instagram')
GROUP BY plataforma;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
tabelas = pd.DataFrame(result,columns=cursor.column_names) # convertendo a variável result em um df pandas chamado tabelas
display(tabelas) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```

```python
'''
Qual plataforma está gerando o maior ROI (Retorno sobre Investimento) para o
segmento "Saúde"?
'''
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database,
) # estas linhas servem para criar a conexão com o banco de dados
cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = '''
SELECT
    p.Nome AS Plataforma,
    SUM(c.Conversoes) / SUM(c.Investimento_Total) AS ROI
FROM
    Campanhas c
INNER JOIN Plataformas p ON c.Plataforma = p.Nome
WHERE c.Segmento_Alvo = 'Saúde'
GROUP BY p.Nome
ORDER BY ROI DESC;
'''
cursor.execute(query) # Execução da consulta
result = cursor.fetchall() # armazenando os resultados na variável results
tabelas = pd.DataFrame(result,columns=cursor.column_names) # convertendo a variável result em um df pandas chamado tabelas
display(tabelas) # exibindo a tabela pandas
cursor.close() # fechando o cursor
connection.close() # fechando a conexão
```
